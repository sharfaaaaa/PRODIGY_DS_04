```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Replace 'your_file.csv' with the actual filename
df = pd.read_csv('twitter.csv', header=None, names=['ID', 'Topic', 'Sentiment', 'Text'])

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Plot the sentiment distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment', data=df, palette='coolwarm')
plt.title('Sentiment Distribution for Borderlands')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Further analysis: Analyze sentiment distribution by topic (if multiple topics exist)
plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment', hue='Topic', data=df, palette='coolwarm')
plt.title('Sentiment Distribution by Topic')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
